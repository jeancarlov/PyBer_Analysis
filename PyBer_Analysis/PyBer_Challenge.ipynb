# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
pd.read_csv(ride_data_to_load, index_col='date', parse_dates=True, infer_datetime_format=True)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_byCity_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_byCity_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_byCity_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_byCity_type
```

```python
#  3. Get the total amount of fares for each city type
total_fare_amount_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fare_amount_type
```

```python
#  4. Get the average fare per ride for each city type. 
# ave_fare_by_type = pyber_data_df.groupby(["city"]).mean()["fare"]
ave_fare_per_ride = total_fare_amount_type / total_rides_byCity_type
ave_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
# avg_fare_per_driver = pyber_data_df.groupby(["city"]).mean()["driver_count"]
avg_fare_per_driver = total_fare_amount_type / total_drivers_byCity_type
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_ride_summary_df = pd.DataFrame({
    "Total Rides": total_rides_byCity_type,
    "Total Drivers": total_drivers_byCity_type,
    "Total Fares": total_fare_amount_type,
    "Average Fare per Ride":ave_fare_per_ride,
    "Average Fare per Driver": avg_fare_per_driver})

pyber_ride_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_ride_summary_df.index.name = None
pyber_ride_summary_df
```

```python
#  8. Format the columns.
pyber_ride_summary_df["Total Rides"] =pyber_ride_summary_df["Total Rides"].map("{:,}".format)
pyber_ride_summary_df["Total Drivers"] =pyber_ride_summary_df["Total Drivers"].map("{:,}".format)
pyber_ride_summary_df["Total Fares"] =pyber_ride_summary_df["Total Fares"].map("{:,}".format)
pyber_ride_summary_df["Average Fare per Ride"] =pyber_ride_summary_df["Average Fare per Ride"].map("{:.2f}".format)
pyber_ride_summary_df["Average Fare per Driver"] =pyber_ride_summary_df["Average Fare per Driver"].map("{:.2f}".format)

pyber_ride_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_fare_type = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
sum_fare_type
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
sum_fare_type = sum_fare_type.reset_index()
sum_fare_type
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
sum_fare_type_pivot = sum_fare_type.pivot(index = "date", columns = "type", values = "fare")
sum_fare_type_pivot
```

```python
# transform regular index to datetime index
sum_fare_type_pivot.index = pd.to_datetime(sum_fare_type_pivot.index)
sum_fare_type_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fare_jan_apr= sum_fare_type_pivot.loc['2019-01-01':'2019-04-29']
fare_jan_apr.head()

```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
fare_jan_apr.index = pd.to_datetime(fare_jan_apr.index)
fare_jan_apr.head()
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
fare_jan_apr.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares_df = fare_jan_apr.resample('W').sum()
weekly_fares_df.head(10)
```
